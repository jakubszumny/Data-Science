```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

print(np.random.random())
```

```python
feature1 = []
for i in range(100):
    feature1.append(np.random.random())


feature2 = []
for i in range(100):
    feature2.append(np.random.random())

df = pd.DataFrame({"Feature1": feature1, "Feature2": feature2})
df["Label"] = 0
df


feature1 = []
for i in range(100):
    feature1.append(np.random.random() + .5)

feature2 = []
for i in range(100):
    feature2.append(np.random.random() + .6)

df2 = pd.DataFrame({"Feature1": feature1, "Feature2": feature2})
df2["Label"] = 1
df2

final_df = pd.concat((df, df2), ignore_index=True)
final_df

colors = []
for i in range(len(final_df)):
    colors.append("C" + str(int(final_df.iloc[i].Label)))
colors

plt.scatter(final_df.Feature1, final_df.Feature2, c = colors)
```

```python
final_df.loc[len(final_df)] = [1.1, .7, 2]
```

```python
colors = []
for i in range(len(final_df)):
    colors.append("C" + str(int(final_df.iloc[i].Label)))

plt.scatter(final_df.Feature1, final_df.Feature2, 
    c = ["C" + str(int(color)) for color in final_df.Label])
```

```python
test_index = final_df.index[-1]
final_df.loc[test_index, "Feature1"] = .3
final_df.loc[test_index, "Feature2"] = 1.4

plt.scatter(final_df.Feature1, final_df.Feature2, c = colors)
```

```python
test_row = final_df.iloc[-1]
test_row
train_df = final_df.iloc[:-1]
train_df

closest_dist = 99999999
closest_label = -1

for i in range(len(train_df)):
    train_row = train_df.iloc[i]
    dist = (train_row.Feature1 - test_row.Feature1) ** 2 + \
        (train_row.Feature2 - test_row.Feature2) ** 2
    if dist < closest_dist:
        closest_dist = dist
        closest_label = train_row.Label

print("Closest Distance:", np.sqrt(closest_dist))
print("Predicted Label:", closest_label)
```
