```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/kubas/Downloads/car.data", engine = "python", names = ['price', "maintenance", "doors", "persons", "trunk", "safety", "acceptance"])
df
```

```python
for i in range(len(df)):
    if df.price[i] == "vhigh":
        df.price[i] = 4
    elif df.price[i] == "high":
        df.price[i] = 3
    elif df.price[i] == "med":
        df.price[i] = 2
    elif df.price[i] == "low":
        df.price[i] = 1


    if df.maintenance[i] == "vhigh":
        df.maintenance[i] = 4
    elif df.maintenance[i] == "high":
        df.maintenance[i] = 3
    elif df.maintenance[i] == "med":
        df.maintenance[i] = 2
    elif df.maintenance[i] == "low":
        df.maintenance[i] = 1



    if df.doors[i] == "2":
        df.doors[i] = 1
    elif df.doors[i] == "3":
        df.doors[i] = 2
    elif df.doors[i] == "4":
        df.doors[i] = 3
    elif df.doors[i] == "5more":
        df.doors[i] = 4


    if df.persons[i] == "2":
        df.persons[i] = 1
    elif df.persons[i] == "4":
        df.persons[i] = 2
    elif df.persons[i] == "more":
        df.persons[i] = 3


    if df.trunk[i] == "small":
        df.trunk[i] = 1
    elif df.trunk[i] == "med":
        df.trunk[i] = 2
    elif df.trunk[i] == "big":
        df.trunk[i] = 3
    

    if df.safety[i] == "low":
        df.safety[i] = 1
    elif df.safety[i] == "med":
        df.safety[i] = 2
    elif df.safety[i] == "high":
        df.safety[i] = 3
    
      
```

```python
df
```

```python
from sklearn import neighbors, model_selection, metrics


features = ["price", "maintenance", "doors", "persons", "trunk", "safety"]

model = neighbors.KNeighborsClassifier(5)
xval = model_selection.KFold(10, shuffle = True)
scorer = metrics.make_scorer(metrics.cohen_kappa_score)
result = model_selection.cross_validate(model, df[features], df.acceptance, cv = xval, scoring = scorer, return_train_score = True)

```

```python
print("Kappa Mean:", result["test_score"].mean())
print("Kappa SD:", result["test_score"].std())
print("Train Kappa Mean:", result["train_score"].mean())
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(min_samples_leaf = 10)
xval = model_selection.KFold(5, shuffle = True)
scorer = metrics.make_scorer(metrics.cohen_kappa_score)
result = model_selection.cross_validate(model, df[features], df.acceptance, cv = xval, scoring = scorer, return_train_score = True)
```

```python
print("Kappa Mean:", result["test_score"].mean())
print("Kappa SD:", result["test_score"].std())
print("Train Kappa Mean:", result["train_score"].mean())
```

```python
from sklearn import ensemble

model = ensemble.RandomForestClassifier(min_samples_leaf= 10)
xval = model_selection.StratifiedKFold(4, shuffle = True)
scorer = metrics.make_scorer(metrics.cohen_kappa_score)
result = model_selection.cross_validate(model, df[features], df.acceptance, cv = xval, scoring = scorer, return_train_score=True)
```

```python
print("Kappa Mean: ", result["test_score"].mean())
print("Kappa SD: ", result["test_score"].std())
print("Train Kappa Mean: ", result["train_score"].mean())
```
